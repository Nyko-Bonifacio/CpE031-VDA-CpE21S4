1.   What is the  SleepHours of people get depending on their Physical Health Day?
2.   What is the average SleepHours of people get depending on their BMI?


```python
#1. What is the SleepHours of people get depending on their Physical Health Day?
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('heart_2022_no_nans.csv')

# Select only the first 10 records
data_first_10 = data.head(10)

days = data_first_10['PhysicalHealthDays']
sleep = data_first_10['BMI']

plt.bar(sleep, days)
plt.ylabel("Physical Health Days")
plt.xlabel("BMI")
plt.title("Sleep Hours of the First 10 Records")
plt.show()
```

```python
#2.  What is the average SleepHours of people get depending on their BMI?
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('heart_2022_no_nans.csv')

# Select only the first 50 records
data_first_50 = data.head(50)

days = data_first_50['BMI']
sleep = data_first_50['SleepHours']

plt.bar(sleep, days)
plt.ylabel("BMI")
plt.xlabel("Sleep Hours")
plt.title("Sleep Hours of the First 50 Records")
plt.show()
```
